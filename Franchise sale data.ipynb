```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\SIDDHESH\Desktop\filter.csv')
df
```

```python
df.drop('Unnamed: 0',axis = 1,inplace=True)
```

```python
df
```

```python
df['sales']=df['Gross Value']-df['Standard Discount']-df['1st Discount Given']
```

```python
df.columns
```

```python
zone_info=df.groupby(['Zone'])['Won/Lost/In Process'].value_counts()
zone_info
```

## Zone wise sales

```python
s=df.groupby(['Zone'])['sales'].sum()
```

```python
z1=df.groupby(['Zone'])['Client expectation Price Bracket'].value_counts()
z1
```

```python
df.sales
```

```python
s=df.groupby(['Zone'])['sales'].sum()

s=s/1000000
zone_wise=s.round(2)
zone_wise.sort_values(ascending=False)
```

```python
plt.figure(figsize=(12,6))
plt.bar(zone_wise.index,zone_wise.values)
plt.xlabel('Zone',fontsize=24)
plt.ylabel("sale in millions",fontsize=24)
for i,j in enumerate(zone_wise):
    plt.text(i,j,str(j),ha='center')
```

## Client expectation 

```python
pref=df['Client expectation Price Bracket'].value_counts()
pref
```

```python
pref=df.groupby(['Client expectation Price Bracket'])['sales'].sum()
pref=pref/1000000
pref.round(2)
```

```python
plt.figure(figsize=(12,5))
colors=plt.cm.Paired(range(len(pref)))
plt.bar(pref.index,pref.values,color=colors)
plt.title('Client expectation Price Bracket',ha="center",fontsize=16)
for i,j  in enumerate(pref):
    plt.text(i,j,str(j))
plt.show()
```

## Won

```python
won=df.loc[df['Won/Lost/In Process']=='Won']
won=won.groupby(['Sales Owner'])['sales'].sum().sort_values(ascending=False)
```

```python
won=(won/1000000).round(2)
```

```python
won
```

```python
top=won.head(10)
top
```

```python
plt.figure(figsize=(12,5))
colors=plt.cm.Paired(range(len(top)))
plt.bar(top.index,top.values,color=colors)
plt.title('top 10 performing in millions',fontsize=16)
for i,j  in enumerate(top):
    plt.text(i,j,str(j))
plt.xticks(rotation=45)
plt.show()
```
